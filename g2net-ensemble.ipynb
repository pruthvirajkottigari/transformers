```python
import gc
import os
import joblib
import numpy as np
import pandas as pd

from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
```

```python
train = [pd.read_csv(f'../input/g2net-prepare-features/train{i}.csv') for i in range(1,8)]
train = pd.concat(train)
train = train[train['target']!=-1]
train.reset_index(inplace = True, drop = True)
train.head()
```

```python
model_lgbm = LGBMClassifier(n_estimators = 40000)
model_cat = CatBoostClassifier(iterations=25, learning_rate=0.5)
model_xgb = XGBClassifier()
# model_svm = SVC()
model_rf = RandomForestClassifier()
```

```python
FEATURES = [i for i in train.columns if i not in ['id', 'target']]
X_train, X_test, y_train, y_test = train_test_split(train[FEATURES], train['target'], test_size=0.25, random_state=42)
```

```python
model_lgbm.fit(X_train,y_train,eval_set=[(X_test,y_test)],early_stopping_rounds=100, verbose=200)
```

```python

model_cat.fit(X_train,y_train,eval_set=[(X_test,y_test)])
```

```python
model_xgb.fit(X_train,y_train,eval_set=[(X_test,y_test)])
```

```python
model_rf.fit(X_train, y_train)
```

```python
predictions = []
test_indexs = []

for j, model in enumerate([model_rf, model_lgbm, model_cat, model_xgb]):
    preds = []
    for i in range(1,81):
        test = pd.read_csv(f'../input/g2net-prepare-features/test{i}.csv')
        preds += [i[1] for i in model.predict_proba(test[FEATURES])]
        if(j == 0):
            test_indexs += test['id'].tolist()
        del test
        gc.collect()
    predictions.append(np.array(preds))
    del preds
    gc.collect()

predictions = np.average(predictions, axis = 0)
sub = pd.DataFrame(list(zip(test_indexs, predictions)), columns = ['id','target'])
sub = sub.groupby('id').mean().reset_index()
sub.to_csv('submission.csv', index = False)
sub.head()
```
